```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.io as pio
pio.templates.default = "simple_white"
```

```python
df = pd.read_csv(
    "https://raw.githubusercontent.com/bdi593/datasets/refs/heads/main/google-trends-celebrities/google-trends-celebrities.csv"
)

df
```

```python
df["datetime"] = pd.to_datetime(df["datetime"], format="%b %Y")
df
```

```python
df["keyword"].unique()
```

### Who has the highest overall average trend index in the entire dataset?


```python
df.groupby("keyword").agg({"trend_index": "mean"}).sort_values(
    "trend_index", ascending=False
).head(5)
```

### Which celebrity reached a trend index of 100 in the very first month of the dataset (Jan 2004)?


```python
df[(df["datetime"] == "2004-01-01") & (df["trend_index"] == 100)]
```

### How many unique celebrities are tracked in this dataset?


```python
df["keyword"].nunique()
```

### In which year did both Amber Heard and Johnny Depp reach their peak trend index of 100?


```python
df_peaks = df[df["trend_index"] == 100]
df_peaks = df_peaks[
    (df_peaks["keyword"] == "Amber Heard") | (df_peaks["keyword"] == "Johnny Depp")
]
df_peaks
```

### Based on the data, who had the higher average trend index during the year 2025?

- Cristiano Ronaldo
- Lionel Messi
- Kylian Mbappe
- Lebron James


```python
rivals_2022 = df[
    (df["datetime"].dt.year == 2025)
    & (
        df["keyword"].isin(
            ["Cristiano Ronaldo", "Lionel Messi", "Kylian Mbappe", "Lebron James"]
        )
    )
]
print(rivals_2022.groupby("keyword")["trend_index"].mean().sort_values(ascending=False))
```

### Filter the following keywords and create a line chart that shows the trend index changes over time for the four keywords.

- Anitta
