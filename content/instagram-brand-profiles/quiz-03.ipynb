```python
import json
import pandas as pd
import glob
```

```python
pd.set_option("display.max_columns", 50)
```

```python
# Load all JSON files
files = glob.glob("data/*.json")

data = []
for file in files:
    with open(file) as f:
        d = json.load(f)
        data.append(d)

# Create DataFrame
df = pd.json_normalize(data)

df
```

### Which brand has the highest number of followers?


```python
df.loc[df["followersCount"].idxmax(), "username"]
```

### What is the average follower count across all brands?


```python
df["followersCount"].mean()
```

### Which brand has posted the most content?


```python
df.loc[df["postsCount"].idxmax(), "username"]
```

### What is National Geographic (@natgeo)'s external URL ("externalUrls")?


```python
df[df["username"] == "natgeo"]["externalUrls"]
```

### What is the average number of likes for McDonald's (@mcdonalds) latest posts?


```python
mcdonalds_latest_posts = df.loc[df["username"] == "mcdonalds", "latestPosts"].iloc[0]
df_mc_latest_posts = pd.DataFrame(mcdonalds_latest_posts)
df_mc_latest_posts
```

```python
df_mc_latest_posts["likesCount"].mean()
```
